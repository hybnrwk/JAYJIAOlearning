# Problem 2.9 误差序列作图实验

这个 Notebook 按课堂示例的结构组织：先设置公共绘图环境，再定义误差序列，最后分别绘制普通坐标图和 semilogy 图。

```
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

BASE = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
PICS = BASE / "pics"
PICS.mkdir(exist_ok=True)

plt.rcParams["figure.dpi"] = 120
plt.rcParams["savefig.dpi"] = 220
plt.rcParams["axes.unicode_minus"] = False
plt.rcParams["font.sans-serif"] = ["Microsoft YaHei", "SimHei", "DejaVu Sans"]

print(f"图片会保存到: {PICS}")
```

## 定义误差序列与收敛速率汇总

迭代下标从 \(k=1\) 开始，避免 \(100/k\) 和 \(1/k^2\) 在 \(k=0\) 处无定义。

```
def make_error_sequences(max_iter=100):
    """Return four error sequences in Problem 2.9.

    The iteration index starts from k=1 because the formulas contain 100/k
    and 1/k^2.
    """
    k = np.arange(1, max_iter + 1, dtype=float)

    errors = {
        r"(a) $4^{-\lfloor k/2\rfloor}$": 1.0 / (4.0 ** np.floor(k / 2.0)),
        r"(b) $\sqrt{5}/2^k$": np.sqrt(5.0) / (2.0 ** k),
        r"(c) $100/k$": 100.0 / k,
        r"(d) piecewise": np.where(k <= 50, 1.0 / (k ** 2), 0.895 ** k),
    }
    return k, errors


def summarize_error_sequences():
    """Summary table for complexity and convergence rate."""
    rows = [
        {
            "sequence": r"$4^{-\lfloor k/2\rfloor}$",
            "iteration_complexity": r"$O(\log(1/\varepsilon))$",
            "rate": r"linear / R-linear, about $O((1/2)^k)$",
        },
        {
            "sequence": r"$\sqrt{5}/2^k$",
            "iteration_complexity": r"$O(\log(1/\varepsilon))$",
            "rate": r"linear, factor $1/2$",
        },
        {
            "sequence": r"$100/k$",
            "iteration_complexity": r"$O(1/\varepsilon)$",
            "rate": r"sublinear, $O(1/k)$",
        },
        {
            "sequence": r"$1/k^2$ if $k\le 50$, $0.895^k$ if $k>50$",
            "iteration_complexity": r"$O(\log(1/\varepsilon))$",
            "rate": r"asymptotically linear, factor $0.895$",
        },
    ]
    return pd.DataFrame(rows)


max_iter = 100
k, errors = make_error_sequences(max_iter=max_iter)
summary_df = summarize_error_sequences()
summary_df
```

## 普通比例尺作图

```
fig, ax = plt.subplots(figsize=(8.5, 5.0))

for label, err in errors.items():
    ax.plot(k, err, linewidth=1.8, label=label)

ax.set_xlabel("iteration k")
ax.set_ylabel(r"error $e(x^k)$")
ax.set_title("Problem 2.9: error sequences, normal scale")
ax.grid(True, alpha=0.3)
ax.legend(fontsize=8)

fig.tight_layout()
out = PICS / "problem_2_9_error_sequences_normal.png"
fig.savefig(out, bbox_inches="tight")
print(f"saved to {out}")
plt.show()
```

## semilogy 作图

这里仅将纵轴改为对数尺度，便于观察线性收敛序列在半对数图中的近似直线形态。

```
fig, ax = plt.subplots(figsize=(8.5, 5.0))

for label, err in errors.items():
    ax.semilogy(k, err, linewidth=1.8, label=label)

ax.set_xlabel("iteration k")
ax.set_ylabel(r"error $e(x^k)$")
ax.set_title("Problem 2.9: error sequences, semilogy scale")
ax.grid(True, which="both", alpha=0.3)
ax.legend(fontsize=8)

fig.tight_layout()
out = PICS / "problem_2_9_error_sequences_semilogy.png"
fig.savefig(out, bbox_inches="tight")
print(f"saved to {out}")
plt.show()
```
